# python opencv 画图相关

```python
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt 
import os
# cv2.imshow('img', img);cv2.waitKey(0);cv2.destroyAllWindows()

os.chdir('/home/zpp/Sample_files/accumulate')
```

```python
# 图像显示函数
def img_show(img_input, figsize_=(2,2), cmap_='rgb'):
    img = img_input
    if cmap_ == 'gray':
        plt.figure(figsize=figsize_)
        plt.imshow(img, cmap =cmap_)
        plt.show()
    elif cmap_ == 'rgb':
        plt.figure(figsize=figsize_)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.show()
    else:
        print('Error: cmap_ is gray or rgb')
```

```python
img_read = cv2.imread('read.jpg')
img_show(img_read)
```

```python
img = np.zeros((512,512,4),np.uint8)
img_show(img)
```

```python
#　画直线，参数为绘图窗口，起点，终点，颜色，粗细，连通性
img_line = cv2.line(img,(100,0),(511,511),(255,0,0),5,8)
img_show(img_line)
```

```python
# 画矩形，参数为窗口，左上角坐标，右下角坐标，颜色，粗度
img_rectangle = cv2.rectangle(img,(150,200),(300,400),(0,255,0),5)
img_show(img_rectangle)
```

```python
#画圆函数，参数为窗口，圆心，半径，颜色，粗度（如果粗度为-1标示实心）,连通性
img_circle = cv2.circle(img,(150,200),50,(0,0,255),5,8)
img_show(img_circle)
```

```python
# 画椭圆函数，参数为窗口，椭圆中心，椭圆长轴短轴，椭圆逆时针旋转角度，椭圆起绘制起始角度，椭圆绘制结束角度，颜色，粗度，连通性
img_ellipse = cv2.ellipse(img,(150,200),(100,50),100,36,360,(0,255,255),5,8)
img_show(img_ellipse)
```

```python
#画多边形函数，参数为窗口，坐标，是否封闭，颜色
pts = np.array([[100,50],[200,300],[400,200],[400,100]], np.int32)
img_polylines = cv2.polylines(img,[pts],True,(255,255,255),5)
img_show(img_polylines)
```

```python
#画多边形函数，参数为窗口，坐标，是否封闭，颜色
pts = np.array([[150,100],[250,350],[450,250],[450,150]], np.int32)
img_polylines = cv2.polylines(img,[pts],False,(255,255,255),5)
img_show(img_polylines)
```

```python
# 画填充多边形
pts = np.array([[300,400],[400,350],[500,500],[300,450]], np.int32)
img_polylines = cv2.fillPoly(img,[pts],(0,255,0))
img_show(img_polylines)
```

```python
# 写字 窗口 ,要写到窗口上的字符 ,字体位置，左下角,字体类型，查询cv2.putText()函数的文档来查看字体 ,字体大小,常规参数，类似颜色，粗细，线条类型等等
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,'ZPP',(10,500), font, 4,(255,255,255),2,cv2.LINE_AA)
img_show(img)
```
